## **Introducción teórica: Machine Learning, modelos que aprenden**

Los algoritmos de machine learning se caracterizan por resolver una **tarea** a partir del aprendizaje mediante **experiencia**. Para ello, resulta fundamental definir una **métrica** que permita evaluar qué tanto aprendió el modelo al enfrentarse a la tarea en cuestión.

En todo proyecto de *machine learning* es necesario comenzar estableciendo cuál es el problema o la tarea a resolver, y de qué manera el modelo puede intervenir para facilitar el logro del objetivo planteado. En términos generales, todo problema de machine learning consiste en procesar un ejemplo de entrada (que puede ser un conjunto de datos) y generar, a partir de él, una salida.

Dentro de las aplicaciones de machine learning, que buscan resolver tareas demasiado complejas para abordarlas con reglas fijas, dos de los problemas más frecuentes son:

*   **Problemas de regresión**
*   **Problemas de clasificación**

Un **problema de regresión** consiste en asignar un valor numérico a una instancia en función del análisis de los datos de *entrada* disponibles. Un ejemplo clásico en la práctica es predecir la cantidad de unidades que venderá un negocio el mes próximo, teniendo en cuenta el historial de ventas y las condiciones bajo las cuales se realizaron.

Un **problema de clasificación**, en cambio, busca asignar una instancia a una clase o categoría según los datos de *entrada* disponibles. Un caso cotidiano es el filtro de correo electrónico, que debe decidir si un mail recibido debe clasificarse como spam o ir a la bandeja de entrada.

La diferencia central entre regresión y clasificación radica en el tipo de salida: en la regresión es un número (generalmente real), mientras que en la clasificación es una de *k* categorías posibles (que eventualmente podrían codificarse con números, pero no lo son por naturaleza).

### **Métricas**
Como se mencionó anteriormente, los modelos de *machine learning* necesitan de una métrica cuantitativa de su performance (eficacia).

- En problemas de *clasificación*, las métricas más utilizadas son:
  - **Accuracy (Precisión global)**: porcentaje de predicciones correctas sobre el total de instancias. Útil cuando las clases están balanceadas.

  - **Precision (Precisión por clase)**: proporción de verdaderos positivos respecto a todas las predicciones positivas. Importante cuando queremos minimizar falsos positivos.

  - **Recall (Sensibilidad o Exhaustividad)**: proporción de verdaderos positivos respecto a todos los positivos reales. Importante cuando queremos minimizar falsos negativos.

  - **F1-Score**: media armónica entre precision y recall.Útil cuando hay un balance entre precisión y recall, especialmente en clases desbalanceadas.

  - **Matriz de confusión**: tabla que muestra cómo se clasifican las instancias entre las distintas clases. Permite ver errores específicos (falsos positivos y falsos negativos). A continuación, se observa una matriz de confusión para una decisión binaria, junto con los cálculos correspondientes a las otras métricas.

- En problemas de *regresión* existen numerosas métricas, las cuáles en general involucran alguna manera de representar la diferencia numérica entre los valores de salida esperados y los predichos. Entre las más conocidas se encuentran el **Mean Absolute Error (MAE), el Mean Squared Error (MSE), el Root Mean Squared Error (RMSE) y el R² (Coeficiente de determinación)**.

  Quizá la más utilizada sea el **Mean Squared Error (MSE)**, cuya fórmula muestra que se suman los cuadrados de las diferencias entre la salida esperada y la predicha:

  $$MSE = \frac{1}{n} \sum^{N}_{j = 1} (y_{j} - \hat{y_{j}})^{2}$$

  Un menor MSE indica un mejor desempeño del modelo.

Estas son algunas de las métricas disponibles, pero no las únicas. Es importante destacar que, para cada problema, la elección de una métrica adecuada al comportamiento deseado puede ser compleja. Esto se debe a que diferentes métricas reflejan distintos aspectos del desempeño del modelo: algunas penalizan más los errores grandes, otras dan igual peso a todos los errores, y otras están diseñadas para tareas específicas, como clasificación con clases desbalanceadas o predicciones numéricas con distintos rangos. Por lo tanto, seleccionar la métrica correcta implica considerar tanto el tipo de problema como el objetivo final del modelo, y priorizar lo que realmente importa medir en la práctica.

### **Datasets**

#### **Aprendizaje supervisado vs. no supervisado**

Para poder aprender, los algoritmos de *machine learning* necesitan un conjunto de *ejemplos* sobre los cuales entrenar. A dicho conjunto lo llamamos **dataset**.  

Los algoritmos de ML pueden clasificarse en dos grandes tipos:  

- **Aprendizaje supervisado**: cada ejemplo del dataset posee un **label** o **target**, que representa la salida real o esperada.  
  - Ejemplo: el *label* puede indicar la categoría a la que pertenece una imagen (*gato*, *perro*, etc.) o el valor numérico asociado a una variable (el precio de una vivienda, la temperatura, etc.).  

- **Aprendizaje no supervisado**: los ejemplos del dataset **no poseen un label o target asociado**.  
  - En estos casos, el algoritmo debe encontrar por sí mismo patrones, relaciones o estructuras dentro de los datos.  
  - Ejemplo: el **clustering**, que consiste en dividir el dataset en grupos basados en características similares, sin conocer de antemano cuáles deberían ser esos grupos.  

**Comentario adicional**:  
Existe también el **aprendizaje por refuerzo**, donde un agente aprende a tomar decisiones interactuando con un entorno y recibiendo recompensas o castigos según sus acciones. Aunque no lo profundizaremos en esta clase, vale la pena mencionarlo ya que es muy utilizado en robótica, videojuegos y sistemas de optimización.

**Calidad de los datos**  
Como en todo proceso de análisis, los resultados que el modelo pueda devolver dependerán en gran medida de la **calidad del dataset**. Algunos aspectos clave son:  

- Asegurar un formato uniforme de los datos.  
- Manejar valores faltantes de forma adecuada.  
- Tratar los *outliers* o valores atípicos para evitar que confundan al modelo.  
- Contar con un volumen de datos suficiente que represente la diversidad de la población real.  

En definitiva, un modelo no podrá predecir más allá de aquello con lo que fue entrenado.  


#### **Datos: Train y Test**

El desafío principal en *machine learning* es lograr que el modelo funcione bien en **datos nuevos**, nunca antes vistos durante el proceso de aprendizaje. En otras palabras, se busca que el modelo **generalice** correctamente y no simplemente que memorice los datos con los que fue entrenado.  

Para evaluar este aspecto, es necesario dividir el **dataset** en distintos subconjuntos:  

- **Datos de entrenamiento (Train)**: son los ejemplos destinados a que el modelo aprenda de ellos. Se utilizan para ajustar los parámetros internos del modelo.

- **Datos de prueba (Test)**: se reservan para la evaluación final del modelo. Permiten verificar cómo se desempeña el modelo con datos completamente nuevos y nunca utilizados en ninguna etapa previa.  

Es fundamental que los dos conjuntos sean **independientes** entre sí, para que la evaluación del modelo sea confiable.

#### **Hiperparámetros y set de Validación**

Los modelos de *machine learning* poseen un conjunto de configuraciones que controlan el comportamiento del algoritmo, pero que **no se ajustan automáticamente** durante el proceso de aprendizaje. A estas configuraciones se las denomina **hiperparámetros**.  

El objetivo es elegir el conjunto de hiperparámetros que brinde el mejor desempeño para el problema. Podría parecer tentador utilizar los datos de *test* para probar distintas configuraciones, pero eso sería incorrecto: al hacerlo, estaríamos filtrando información de dichos datos hacia el modelo de manera indirecta, perdiendo su rol como medida objetiva e imparcial de rendimiento.  

Por este motivo, se utiliza un **set de validación**. En la práctica, el conjunto original de entrenamiento se subdivide en:  

- **Train**: con el cual el modelo aprende ajustando sus parámetros internos.  
- **Validación**: con el cual se evalúa el desempeño del modelo y se ajustan los hiperparámetros.  

Una división común es asignar aproximadamente un **80% de los datos para entrenamiento** y un **20% para validación** (aunque estos porcentajes pueden variar según el problema y la cantidad de datos disponible).  

De esta forma, es posible detectar problemas como el *overfitting* y validar la efectividad del modelo con datos distintos de los usados directamente en el entrenamiento.


### **Overfitting y Underfitting**

Al entrenar un modelo de *machine learning* buscamos que pueda **aprender de los datos** y luego generalizar correctamente a nuevos casos. Sin embargo, durante este proceso pueden aparecer problemas que afectan su rendimiento:

- **Underfitting**: aparece cuando el modelo es **demasiado simple** para capturar la estructura de los datos.  
  - Se traduce en un **mal desempeño tanto en entrenamiento como en validación**.  
  - El modelo no logra reconocer patrones relevantes ni relaciones entre variables.  
  - Para corregirlo, se puede:  
    - Seleccionar mejores variables predictoras.  
    - Incrementar la complejidad del modelo (más parámetros, arquitecturas más expresivas).

- **Overfitting**: en cambio, ocurre cuando el modelo aprende en exceso los detalles y el ruido de los datos de entrenamiento.  
  - Se observa un **alto desempeño en el entrenamiento**, pero un **bajo desempeño en validación**.  
  - En lugar de identificar patrones generales, el modelo termina **“memorizando” los datos**.  
  - Algunas formas de mitigarlo incluyen:  
    - Aumentar la cantidad o diversidad de datos.  
    - Reducir la complejidad del modelo (menos parámetros, regularización, poda, etc.).

### **Síntesis**

Es importante tener en mente que las inteligencias artificiales, por lo general, **no son más que grandes bases de datos** sobre las cuales se realiza un análisis para generar respuestas o predecir resultados. Por lo tanto, la **precisión de la IA**, es decir, qué tan acertada es, depende directamente de **los datos que se utilizan** y de **cómo los interpreta**.

#### Ejemplo práctico

Supongamos que creamos una IA para **distinguir y clasificar imágenes de manzanas y naranjas**. Cuando se le presente una imagen de alguna de estas frutas, el programa debe indicar correctamente cuál es.  

- **Caso 1: Base de datos pequeña**  
  Si la base de datos contiene solo **3 imágenes de cada fruta**, el modelo **no tendrá suficientes ejemplos** para aprender patrones confiables. Por ejemplo, si se le muestra una manzana con tonos más amarillos que rojos, probablemente la confunda con una naranja.  

- **Caso 2: Base de datos grande y variada**  
  Si la base de datos incluye **miles de imágenes con características variadas**, el modelo podrá **reconocer correctamente el tipo de fruta** incluso si la imagen no coincide exactamente con los rasgos “estándar” de cada fruta.  

**Conclusión:** La capacidad de una IA para generalizar y acertar depende **fundamentalmente de la cantidad y calidad de los datos** con los que fue entrenada.

Comenzaremos ejemplificando un **problema de regresión**.

#### **Ejemplo práctico**

Supongamos que queremos analizar dos variables relacionadas, como el **peso** y la **altura** de una persona.  
- En este caso, sería útil contar con una **función que vincule una variable con la otra**, de modo que podamos estimar el valor de una a partir del conocimiento de la otra.  
- Esta lógica se aplica a cualquier situación en la que queramos predecir valores de variables relacionadas usando solo algunos datos conocidos.  

El desafío principal consiste en determinar **cómo es la función que vincula las variables**. Existen diversos métodos para proponer una función adecuada, desde simples relaciones lineales hasta modelos más complejos de *machine learning*.  

Antes de elegir un método, es importante **examinar los datos** con los que vamos a trabajar. Para nuestro ejercicio, consideraremos como condición inicial las dos variables mencionadas: **peso** y **altura**.

En este caso, nuestro *dataset* consistirá de un conjunto de ejemplos, cada uno de los cuales tendrá como dato de *entrada* a la variable **altura** y como salida a la variable **peso**. El dataset cuenta, para cada ejemplo, con tanto los valores de entrada como los de salida correspondientes (los *targets*), por lo que este problema se podría considerar un problema de *aprendizaje supervisado* y de *regresión*.

```python
'''Consideremos el caso en que tenemos una serie de alturas asociadas
a un determinado peso por tener la misma posición en sus respectivas listas,
en ese caso podemos graficar ambos valores en base a lo ya visto'''

import matplotlib.pyplot as plt
Alturas = [180, 162, 168, 175, 193, 187, 157, 176, 169, 163, 200, 172, 182]
Pesos = [100, 70, 74, 82, 110, 95, 57, 73, 72, 71, 120, 84, 90]
plt.scatter(Alturas, Pesos, color='blue', marker='o')
plt.title("Relación entre altura y peso")
plt.xlabel("Altura (cm)")
plt.ylabel("Peso (kg)")
plt.grid(True)
plt.show()
```

Se observa de forma inmediata que existe una relación entre ambas variables: a mayor altura de una persona, mayor tiende a ser su peso, y viceversa.

El problema ahora radica en cómo expresar cuantitativamente esta relación, ya que simplemente afirmar que un valor aumenta con el otro no es suficiente para realizar predicciones.

### **Regresión Lineal**

Para abordar este problema existen diversos métodos. El primero que vamos a analizar —por ser el más sencillo y directo— es la **regresión lineal**. Es decir, dada una serie de datos que vinculan dos variables (como en el ejemplo provisto), buscamos cuál es la recta que mejor se ajusta a esos valores.

En Python, esto puede hacerse utilizando el módulo **numpy** y la función **np.polyfit**. A esta función hay que darle los datos utilizados para el análisis y nos devuelve los coeficientes de la función lineal obtenida. Recordar que toda recta  puede expresarse en términos generales como $y = x \cdot a + b$ siendo a y b números.

El modelo entonces consiste en una simple cuenta, para la cual, a partir de los datos disponibles, habrá que hallar los valores de $a$ y $b$ que mejor se ajusten al problema. Para hacer esto, la función *polyfit* calcula cuáles son los valores de $a$ y $b$ que minimizan el $MSE$ (error cuadrático medio), mencionado en las secciones anteriores.

 A continuación, graficando dicha recta, se puede ver cómo resulta la aproximación obtenida:


```python
import matplotlib.pyplot as plt
import numpy as np

Alturas = [180, 162, 168, 175, 193, 187, 157, 176, 169, 163, 200, 172, 182]
Pesos = [100, 70, 74, 82, 110, 95, 57, 73, 72, 71, 120, 84, 90]

a, b = np.polyfit(Alturas, Pesos, 1)

X = []
Y = []

for x in range(min(Alturas), max(Alturas) + 1):
    X.append(x)
    y = a * x + b
    Y.append(y)

plt.scatter(Alturas, Pesos, color='blue', marker='o', label='Datos')
plt.plot(X, Y, color='red')
plt.title("Regresión lineal")
plt.xlabel("Altura (cm)")
plt.ylabel("Peso (kg)")
plt.grid(True)
plt.show()
```

De lo anterior se puede ver que la función lineal propuesta se asemeja considerablemente a los datos reales brindados por el ejercicio, aunque la similitud no es perfecta.

La ventaja, sin embargo, de tener esta función es que ahora podemos estimar el peso de una persona con solo conocer su altura, o estimar su altura a partir del peso. Esto permite realizar predicciones relativamente acertadas, incluso sin contar con toda la información original usada para ajustar la función.

### **Regresión polinómica**


Para responder a las primeras preguntas, es importante notar que la función np.polyfit no se limita a la regresión lineal, sino que puede utilizarse para ajustar un polinomio de cualquier grado a los datos proporcionados. El grado del polinomio se especifica como el tercer parámetro de la función (por ejemplo, 1 para regresión lineal).

Retomando el planteo anterior, ahora podemos aplicar una función cuadrática ajustando un polinomio de grado 2:

```python
import matplotlib.pyplot as plt
import numpy as np

Alturas = [180, 162, 168, 175, 193, 187, 157, 176, 169, 163, 200, 172, 182]
Pesos = [100, 70, 74, 82, 110, 95, 57, 73, 72, 71, 120, 84, 90]

a, b, c = np.polyfit(Alturas, Pesos, 2)

X = []
Y = []
for x in range(min(Alturas), max(Alturas) + 1):
    X.append(x)
    y = a * x**2 + b * x + c
    Y.append(y)

plt.scatter(Alturas, Pesos, color='blue', marker='o', label='Datos')
plt.plot(X, Y, color='red')
plt.title("Regresión Cuadrática")
plt.xlabel("Altura (cm)")
plt.ylabel("Peso (kg)")
plt.grid(True)
plt.show()
```

Normalmente, aumentar el grado del polinomio tiende a mejorar la capacidad del modelo para aproximar los datos, ya que le otorga mayor flexibilidad a la curva. Sin embargo, esto implica un costo importante: el riesgo de **sobreajuste (overfitting)**. ### **Naive Bayes**

#### **Bases de probabilidad**

Como ya hemos mencionado, existen otros métodos para realizar predicciones a partir de datos previamente conocidos. En particular, ahora vamos a hablar acerca de un método útil para resolver problemas de clasificación en lugar de regresión como el caso anterior. Este método es el de Naive Bayes (Bayes Ingenuo) y se basa fuertemente en el uso de la **probabilidad**.

Aunque este curso no tiene como objetivo brindar una explicación completa sobre teoría de la probabilidad, es necesario saber ciertos conceptos para comprender cómo funciona este método. A continuación, se presentan algunas reglas fundamentales de la probabilidad que serán necesarias para entender el desarrollo posterior:

* La suma de la probabilidad de que un evento ocurra y de que no ocurra es igual a 1, es decir, sea A un evento cualquiera entonces:

<br>
$$
P(A) + P(\neg A) = 1
$$
<br>

Esto ocurre porque un evento puede ocurrir o no ocurrir, no existe una tercera opción. Por tanto, la suma de las probabilidades de ambas opciones debe dar una posibilidad de 1, es decir, un evento que siempre ocurre.

* La probabilidad de que ocurra un evento A dado que ocurra otro evento B viene dada por el teorema de Bayes:

  
$$
P(A|B) = \frac{P(B|A)P(A)}{P(B)}
$$
  

**Importante**: por notación en probabilidad, $P(A|B)$ se lee como "probabilidad de que ocurra A dado que ocurra B". Por dar un ejemplo, se podría decir que la probabilidad de que alguien falte al trabajo si se encuentra enfermo es del 50%. Escrito en formato de ecuación:

<br>  
$$
P(\text{Faltar al trabajo} \mid \text{Estar enfermo}) = 0.5
$$
<br>

* La probabilidad de que ocurran eventos independientes viene dada por el producto de que ocurra cada evento por separado:

<br>
$$
P(\text{A y B}) = P(\text{A}) P(\text{B})
$$
<br>

Por ejemplo, la posibilidad de tirar una moneda y de que salga cara dos veces es el producto de que salga cara una vez y que salga cara la segunda; para cada tiro se tiene un 50% de que salga cara o cruz al ser una moneda "justa". Por tanto, la probabilidad resulta:

<br>
$$
P(\text{Cara y Cara}) = P(\text{Cara}) P(\text{Cara}) = 0.5 \cdot 0.5 = 0.25 = 25\text{%}
$$
<br>










#### **Explicación del método**

Entonces, habiendo explicado brevemente algunos conceptos básicos de probabilidad, podemos analizar cómo funciona este modelo. Para dicho análisis utilizaremos un ejemplo ya mencionado: detectar si un mensaje es spam o no lo es.

Naive Bayes es un modelo que usa el principio **Maximun a Posteriori** para realizar la clasificación. Esto quiere decir que, para el caso de Spam vs No Spam, busca que luego de saber cómo está compuesto el mensaje del mail, cuál es la clase más probable a la que pertenezca el mensaje (si Spam o no Spam). Es decir, es un modelo que usa probabilidades estimadas a partir de los datos. Matemáticamente, dado que se conoce el mensaje $M$, busca la clase $C$ (Spam o No Spam), que maximice :

$$​P(C ∣ \text{mensaje})$$

Vamos a suponer que contamos con los siguientes datos iniciales:

*   Se tienen 1000 mensajes recibidos, de estos:
    *   700 son mensajes normales. De estos:
        *   50 tienen la palabra "dinero"
        *   400 tienen la palabra "buenos"
        *   30 tienen la palabra "tarjeta"
        *   320 tienen la palabra "saludos"
        (total de 800 palabras)
    *   300 son mensajes spam. De estos:
        *   200 tienen la palabra "dinero"
        *   150 tienen la palabra "buenos"
        *   60 tienen la palabra "tarjeta"
        *   170 tienen la palabra "saludos"
        (total de 580 palabras)

<br>

|  | "dinero" | "buenos" | "tarjeta" | "saludos" | Total |
|----------|----------|----------|-------|-----|-----|
| Normal | 50 | 400 | 30 | 320 | 700 |
| Spam | 200 | 150 | 60 | 170 | 300 |
|Totales| 250 | 550 | 90 | 490 | 1000 |

A partir de estos resultados, podemos hacer aproximaciones de las probabilidades presentes con respecto a los mensajes recibidos. En primer lugar sobre todos los mensajes tenemos:

<br>  
$$
P(\text{Mensaje es normal}) = 700/1000 = 0.7
$$
$$
P(\text{Mensaje es spam}) = 300/1000 = 0.3
$$
<br>

A su vez, dentro de los mensajes contamos con las probabilidades de que ciertas palabras aparezcan, dependiendo de si el mensaje es normal o spam. Nuestra idea es calcular la probabilidad de que un mensaje sea Spam o no, dado que aparece alguna/s de las 4 palabras que elegimos analizar. Para esto, podemos aplicar el teorema de Bayes y obtener:

$$P(spam | mensaje) = \frac{P(spam)\cdot P(mensaje|spam)}{P(mensaje)}$$

Donde "mensaje" es la secuencia de palabras recibidas. Ahora, vamos a asumir que todas las palabras son independientes entre sí, dado que conocemos la clase (esta es la asunción que simplifica el cálculo y le da el nombre *naive* a este modelo). Es decir, vamos a asumir que podemos modelar qué tan probable es que aparezcan las palabras que aparecen en el mensaje como el producto de qué tan probable es que aparezca cada palabra, siendo spam. Matemáticamente:

$$P(spam | mensaje) = \frac{P(spam)\cdot P("palabra_1"|spam) \cdot P("palabra_2"|spam)...}{P(mensaje)}$$

La idea es comparar esta cuenta con $P(No Spam | mensaje)$ y quedarse con la que de mayor. En ambos casos, se estará dividiendo por $P(mensaje)$, por lo que, al fin de cuentas, deberemos calcular y comparar estas 2 ecuaciones:


$$
\text{puntaje Spam} = P(\text{spam}) \cdot P(\text{palabra}_1 \mid \text{spam}) \cdot P(\text{palabra}_2 \mid \text{spam}) \cdot \ldots
$$

con la ecuación:

$$
\text{puntaje No Spam} = P(\text{no spam}) \cdot P(\text{palabra}_1 \mid \text{no spam}) \cdot P(\text{palabra}_2 \mid \text{no spam}) \cdot \ldots
$$



Para nuestro ejemplo, tenemos las siguientes probabilidades de ocurrencia de palabras:



<br>  
$$
P(\text{"dinero"} \mid \text{normal}) = \frac{50}{800} = 0.0625
$$

$$
P(\text{"buenos"} \mid \text{normal}) = \frac{400}{800} = 0.5
$$

$$
P(\text{"tarjeta"} \mid \text{normal}) = \frac{30}{800} = 0.0375
$$

$$
P(\text{"saludos"} \mid \text{normal}) = \frac{320}{800} = 0.4
$$
<br>
$$
P(\text{"dinero"} \mid \text{spam}) = \frac{200}{580} \approx 0.345
$$

$$
P(\text{"buenos"} \mid \text{spam}) = \frac{150}{580} \approx 0.259
$$

$$
P(\text{"tarjeta"} \mid \text{spam}) = \frac{60}{580} \approx 0.103
$$

$$
P(\text{"saludos"} \mid \text{spam}) = \frac{170}{580} \approx 0.293
$$
<br>

Una vez obtenidas estas probabilidades, el método simplemente consiste en realizar el cálculo correspondiente en función de las palabras que aparecieron en el mismo, y fijarse qué cuenta da mayor.

Veamos un ejemplo de como funciona la lógica:

<br>
<img src="https://i.imgur.com/Eplpuob.png" width=850>
<br>



**Ejemplo**:

Si aparece el mensaje "Dame dinero por favor, saludos", la cuenta sería:

$$
\text{puntaje Spam} = P(\text{spam}) \cdot P(\text{saludos} \mid \text{spam}) \cdot P(\text{tarjeta} \mid \text{spam}) = 0,3 \cdot 0,345 \cdot 0.293 = 0,0303255
$$


$$
\text{puntaje no Spam} = P(\text{no spam}) \cdot P(\text{buenos} \mid \text{no spam}) \cdot P(\text{tarjeta} \mid \text{no spam}) = 0,7 \cdot 0,0625 \cdot 0.4 = 0,0175
$$

Al comparar ambos resultados, observamos que el mayor es el correspondiente a spam, por lo que se lo clasifica como Spam.

A continuación, podemos ver un código que realiza la cuenta con dichos valores.

```python
def naiveBayes(mensaje):
    Contador = [0.3, 0.7] #Probabilidades iniciales de que el mensaje sea Spam (Contador[0]) o no lo sea (Contador[1])
    palabras = ['dinero', 'buenos', 'tarjeta', 'saludos']
    probDadoSpam = [0.345, 0.259, 0.103, 0.293]
    probDadoNoSpam = [0.0625, 0.5, 0.0375, 0.4]

    mensaje = mensaje.lower().split() #Separamos el mensaje en las palabras individuales y reducimos en caso de mayusculas

    for palabra in palabras:
        count = mensaje.count(palabra)
        i = palabras.index(palabra)
        Contador[0] *= probDadoSpam[i] ** count # Como consideramos que las palabras son independientes la probabilidad final es el producto de las probabilidades individuales
        Contador[1] *= probDadoNoSpam[i] ** count

    print("Puntaje spam:", Contador[0])
    print("Puntaje no spam:", Contador[1])

    if Contador[0] >= Contador[1]:
        return "Es Spam"
    else:
        return "No es Spam"


res = naiveBayes('Dame dinero por favor saludos')
print((res + '\n'))
res = naiveBayes('Buenos dias amigo, como estas?, saludos')
print(res)


```

A continuación, veamos cómo se implementaría Naive Bayes para la detección de mensajes spam, tomando como base el análisis que hicimos anteriormente.

Primero, se utiliza un **CountVectorizer**, que recibe texto, genera un vocabulario y, para cada mensaje, crea un **vector de ocurrencias**. Es decir, convierte cada mensaje en un vector numérico donde se indica cuántas veces aparece cada palabra del vocabulario en ese mensaje. Este paso es fundamental porque los modelos de machine learning no pueden trabajar directamente con texto plano.

Luego, se llama al modelo de **Naive Bayes** y se le pasan estos vectores junto con las etiquetas correspondientes (`"spam"` o `"no_spam"`). Con la función `fit`, el modelo calcula automáticamente todas las probabilidades necesarias:  
- las **probabilidades a priori** de cada clase,  
- y las **probabilidades condicionales** de cada palabra dado cada clase.

A diferencia del ejemplo teórico donde solo considerábamos cuatro palabras, aquí el modelo tiene en cuenta **todas las palabras** presentes en los mensajes de entrenamiento y la **frecuencia** con la que aparecen. Esto permite que el modelo aprenda patrones más precisos, incluso si se trabaja inicialmente con un conjunto pequeño de mensajes. Con un dataset más grande, el modelo puede generalizar mucho mejor.

En esta etapa, también es útil analizar qué palabras conviene incluir en el análisis:  
- si mantener las palabras más frecuentes,  
- si incluir palabras poco comunes que podrían ser discriminativas,  
- o aplicar técnicas de procesamiento de lenguaje natural, como eliminación de stopwords, lematización o filtrado de caracteres especiales.

Una vez realizado el `fit`, se puede proceder a la **predicción** con el método `predict`. Este método:  
- convierte automáticamente el nuevo mensaje a un vector numérico usando el vocabulario aprendido,  
- calcula los **puntajes de cada clase** utilizando las probabilidades aprendidas,  
- y devuelve la clase con mayor probabilidad, aplicando la regla MAP (Maximum a Posteriori).

Finalmente, es importante destacar que Naive Bayes es un modelo **rápido y efectivo**, especialmente en tareas de clasificación de texto, aunque asume que las palabras son independientes entre sí (de ahí el “naive”). Esta suposición simplifica los cálculos y, sorprendentemente, funciona muy bien en la práctica para detección de spam y otras tareas de procesamiento de lenguaje natural.





```python
# Importamos varias herramientas del módulo sklearn
from sklearn.feature_extraction.text import CountVectorizer # Sirve para generar una matriz que nos informa cuantas veces aparece cada palabra en un mensaje
from sklearn.naive_bayes import MultinomialNB # Sirve para hacer la operación de probabilidades ya mencionada, va a multiplicar la probabilidad de que el mensaje sea spam o no spam de por si por las probabilidades de que lo sea dadas las palabras que aparecen

# Le damos Datos de entrenamiento
mensajes = ['gana dinero fácil', 'buenos días amigo', 'obtén tu tarjeta de crédito', 'saludos cordiales', 'Que gusto verte amigo', 'Buenos días, espero estes bien', 'dinero rapido ya']
etiquetas = ['spam', 'no_spam', 'spam', 'no_spam', 'no_spam', 'no_spam', 'spam']

# Convertimos los mensajes a vectores numéricos
vectorizador = CountVectorizer()
X = vectorizador.fit_transform(mensajes)

# Entrenamos al modelo
modelo = MultinomialNB()
modelo.fit(X, etiquetas)

# Clasificamos 4 mensajes
mensajes = ['dame dinero', 'buenos dias, dame dinero', 'buenos dias, como estas?', 'dame tu tarjeta de credito amigo']
for mensaje in mensajes:
  X_nuevo = vectorizador.transform([mensaje])
  prediccion = modelo.predict(X_nuevo)
  print(prediccion[0])

```

Como pueden ver la predicción no es perfecta, el segundo mensaje es claramente spam sin embargo no es registrado como tal. Esto sucede en gran parte porque los datos que le estamos dando al módelo no son demasiados, lo que resulta en que el mismo sea incapaz de realizar predicciones acertadas. A más mensajes que se le brinden como entrenamiento mayor debería ser la precisión asociada.

### **¿Qué es una Red Neuronal?**

#### **El Perceptrón**

Para poder entender cómo funciona una Red Neuronal será necesario introducir al elemento fundamental que constituye a las mismas, que es el **"Perceptrón"** o **"Neurona Artificial"**. Este elemento está inspirado en el funcionamiento biológico de una Neurona.

<img src="https://upload.wikimedia.org/wikipedia/commons/b/b0/Perceptr%C3%B3n_5_unidades.svg" width="400"> <img src="https://upload.wikimedia.org/wikipedia/commons/6/62/Neurona.svg" width="500">

<small>*(Por Alejandro Cartas - Trabajo propio, CC BY-SA 4.0, https://commons.wikimedia.org/w/index.php?curid=41534843)*<small>

<small>*(CC BY-SA 3.0, https://commons.wikimedia.org/w/index.php?curid=1561753)*<small>

A grandes rasgos, un **perceptrón simple** se compone de las siguientes partes:

- **Señales de entrada**: Son los datos de entrada, cada uno representado con un número.
- **Pesos**: son números que multiplican a cada una de las entradas.
- **Sumatoria**: es la etapa que realiza la suma de las entradas ponderadas por sus respectivos pesos.
- **Función de activación**: es una función matemática que permite transformar la salida de la sumatoria. Introducen *no linealidades*, permitiendo a la red aprender patrones complejos.


Esta estructura de un perceptrón puede considerarse como un modelo matemático muy simplificado de una neuronal biológica. Así como las neuronas biológicas reciben impulsos nerviosos de otras neuronas a através de sus "dendritas" y, frente a superar cierto umbral, se envía la señal a lo largo de su "Axón" hasta llegar a la conexión con otras neuronas; los "perceptrones" reciben datos de perceptrones anteriores, los combinan y luego de una función de activación, generan una salida para los siguientes perceptrones.

Cabe aclarar que hoy en día las redes neuronales no buscan imitar fielmente la biología, sino que la inspiración biológica sirvió para dar la idea de “unidades simples conectadas en red que aprenden patrones”.


##### **Funciones de activación más utilizadas**

<img src="https://miro.medium.com/v2/resize:fit:720/format:webp/1*ZafDv3VUm60Eh10OeJu1vw.png" width="800">

Existen diferentes funciones de activación que permiten **mapear la salida de la combinación lineal de las entradas** a distintos rangos de valores. La elección de la función de activación depende de varios factores:  
- la etapa de la red neuronal en la que se aplica (capas ocultas o capa de salida),  
- el tipo de problema a resolver (clasificación, regresión, etc.),  
- y consideraciones computacionales y de estabilidad numérica.

Algunas de las funciones de activación más utilizadas son:

- **Sigmoid**:  
  - Fórmula: $$\sigma(x) = \frac{1}{1 + e^{-x}}$$  
  - Salida en el rango (0, 1).  
  - Muy usada en la capa de salida para **clasificación binaria**.  

- **Tanh (Tangente hiperbólica)**:  
  - Fórmula: $$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$  
  - Salida en el rango (-1, 1).  
  - Útil en capas ocultas cuando se desea centrar los datos alrededor de cero.  

- **ReLU (Rectified Linear Unit)**:  
  - Fórmula: $$\text{ReLU}(x) = \max(0, x)$$  
  - Muy popular en capas ocultas de redes profundas por su eficiencia y por ayudar a mitigar el problema del **gradiente desaparecido**.  

- **Softmax**:  
  - Fórmula: $$\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}$$  
  - Convierte un vector de valores en probabilidades que suman 1.  
  - Muy utilizada en la capa de salida para **clasificación multiclase**.  

> Elegir la función de activación adecuada es clave para que la red aprenda correctamente y para garantizar estabilidad en el entrenamiento.

#### **Entrenamiento y aprendizaje**

Hasta ahora hemos explicado brevemente qué operaciones realiza una Red Neuronal y que es capaz de aprender a realizar tareas de clasificación y regresión complejas. Sin embargo, no hemos explicado cómo es posible que la Red Neuronal aprenda, a partir de los datos disponibles, cómo realizar la tarea para la cual se la quiere utilizar.

En primer lugar, cabe preguntarse *qué es* lo que una red neuronal debe aprender. En general, cuándo se entrena una red neuronal, la arquitectura de la red queda fija, por lo que la forma en qué se conectan las neuronas y sus funciones de activación quedan fijas. Por lo tanto, los parámetros que se terminan aprendiendo y ajustando para el problema en cuestión son los **pesos**, es decir, aquellos números que multiplicaban a las entradas a cada neurona, para luego ser sumadas entre sí.


Ahora bien, para entender cómo se ajustan dichos pesos, presentaremos el siguiente esquema general sobre el proceso de aprendizaje de un modelo de machine learning:


<img src="https://raw.githubusercontent.com/IEEESBITBA/Curso-python/refs/heads/master/_assets/modelo%20aprendizaje.jpg" width="600">

Al igual que otros modelos de Machine Learning, el conjunto de datos disponibles se lo suele dividir en Train y Test. En las redes neuronales, se suele dividir en tres conjuntos: **Train, Validation y Test**. Para realizar el entrenamiento de la Red Neuronal, se utiliza el conjunto de datos de **Train**:

- $\vec{X}$: Es el conjunto de datos de entrada pertenecientes al conjunto de Train.
- $\vec{y}$: Es el conjunto de datos de salida esperados, pertenencientes al conjunto de Train.

El par $(\vec{X}_i, \vec{y}_i)$ es un elemento que se encuentra anotado en el conjunto de Train, de manera tal de que se conoce para cada elemento cuál es la entrada y cuál es la salida que debería obtenerse.

 A continuación, explicaremos en qué consiste cada etapa del proceso de aprendizaje.

 - **Modelo**: en este caso la red neuronal, recibe un conjunto de entradas, y devuelve un conjunto de salidas estimadas $\hat{\vec{y}}$. El modelo depende de los parámetros $\theta$, que en el caso de la red neuronal, son los **pesos** que serán aprendidos.

- **Función de costo**: la idea fundamental es lograr que la salida estimada $\hat{\vec{y}}$ y la salida esperada $\vec{y}$ sean iguales o lo más similares posible, dependiendo el problema. Para esto, es necesario tener una manera de cuantificar qué tan similares son. Es por esto que se define una **función de costo**, que lo que hace es realizar la comparación entre $\hat{\vec{y}}$ y $\vec{y}$ y devolver un número que represente cuánto es el **error**. Cuánto más similares sean, la función de costo será menor.

- **Optimizador**: conociendo cómo es el modelo y sus parámetros (pesos) $\theta$ a aprender, y conociendo cómo es la función de costo y cuánto dio para un conjunto de datos de Train, el optimizador se encarga de modificar los pesos de manera tal de intentar lograr llevar a la función de costo a un mínimo.

La idea es intuitiva, para el conjunto de datos disponible, se evalúa qué tan bien da el modelo actual a través de una función de costo, y luego se actualizan los pesos de manera tal de que ese error sea, idealmente, mínimo.

Existen distintos tipos de función de costo dependiendo el problema en particular a resolver. Entre las más usadas, se encuentran el **Error cuadrático medio (MSE)**, común en problemas de regresión y la **Entropía cruzada (cross-entropy)**, estándar en clasificación.

A su vez, existen diferentes tipos de optimizadores, para mencionar algunos ejemplos, **Gradient Descent**, **Stochastic Gradient Descent (SGD)** y **Adam**. Para obtener una intuición de cómo funcionan en general estos optimizadores, en la siguiente imagen se muestra una representación bidimensional de cómo se podría llegar a visualizar una función de costo. A mayor altura, mayor será la función de costo, por lo que la idea es poder llegar a un mínimo aceptable, y usar los pesos obtenidos en dicho mínimo.


<img src="https://easyai.tech/wp-content/uploads/2019/01/tiduxiajiang-1.png.webp" width="600">

Los optimizadores arrancan, a partir de un conjunto de pesos iniciales, en algún punto de la superficie de la función de costo, y se encargan de calcular matemáticamente hacia qué dirección deben moverse para descender lo más posible. En general, los optimizadores tienen sus bases en el **Gradient Descent**, que lo que hace es calcular el **gradiente** matemático de la función de costo; que no es otra cosa que la dirección en la que más crece (o decrece en el sentido contrario) una función. De esta manera, cada uno con sus propias variantes, los optimizadores van ajustando los pesos hasta llegar a un mínimo, como se observa en el camino de la figura.

El cálculo de los gradientes usa la **regla de la cadena** propagando el error hacia atrás desde la salida hasta las capas iniciales, lo cual podrán encontrar bajo el nombre de **"backpropagation"**.



### **Ejemplo 1: clasificación de conjuntos de puntos en un plano 2D**

Ahora sí, habiendo presentado esta introducción a los fundamentos principales de las redes neuronales, vamos a mostrar cómo es posible utilizar Python para crear y entrenar nuestro propio modelo, para resolver un problema de clasificación de puntos en un plano en 2D.

Para esto, utilizaremos la siguiente librería, muy utilizada en el mundo de las Redes Neuronales:

- **Tensorflow**: es una librería de código abierto desarrollada por Google para construir y entrenar redes neuronales y otros modelos de machine learning. Permite trabajar tanto con modelos simples como con arquitecturas profundas, y aprovecha la GPU para acelerar los cálculos. Documentación oficial de TensorFlow. *Documentación*: https://www.tensorflow.org/?hl=es

En primer lugar, importaremos las librerías que vamos a estar utilizando, junto con algunas funciones que nos serán de utilidad para generar el problema de clasificación y posteriormente encarar una solución con nuestra propia Red Neuronal.

```python
# 1. Librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import make_circles
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

Lo primero que haremos es generar un dataset para este ejemplo. En general, los datasets para un problema particular son un conjunto de datos representativos que deben ser recolectados, almacenados y etiquetados de manera correcta. En este caso, no utilizaremos un dataset pre-existente ni recolectaremos nuestros propios datos del mundo real, sino que generaremos un dataset **sintético** con un conjunto de puntos en un plano 2D, con una distribución de dos círculos concéntricos. Cada punto pertenecerá a dos posibles clases, que visualizaremos con los colores azul y rojo.

Utilizaremos la función [**make_circles**](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_circles.html) de Scikit Learn, generando 1000 puntos en la configuración mencionada, con algún grado de aleatoriedad para que algunos puntos de un color queden del lado donde deberían ir los del otro color.

```python
# 2. Generamos un dataset sintético 2D
X, y = make_circles(n_samples=1000, noise=0.15, factor=0.2, random_state=42)

# Lo convertimos en DataFrame
df = pd.DataFrame(X, columns=['x1', 'x2'])
df['target'] = y

# Información general con pandas
print(f"Dimensiones del DataFrame: {df.shape}\n")
print("Primeros 5 elementos del dataset:")
print(df.head())

print("\nCantidad de muestras por clase:")
print(df['target'].value_counts())
```

Utilizaremos ahora Matplotlib para visualizar a los puntos generados. Los puntos azules vemos que pertenecen a la clase "0" y los puntos rojos a la clase "1". Por lo tanto, el objetivo de la red neuronal será aprender a poder clasificar a los puntos, dadas sus coordenadas, a qué clase pertenecen.

```python
# 3. Visualizamos el dataset
plt.figure(figsize=(8, 6))
plt.scatter(df['x1'], df['x2'], c=df['target'], cmap='bwr', edgecolor='k')
plt.title('Datos para Clasificación Binaria')
plt.colorbar(label='Valor de y')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```

Empezaremos por dividir a nuestros datos disponibles en los conjuntos de Train y Test. El conjunto de Train serán los datos que utilizará la red neuronal para aprender durante el entrenamiento los patrones necesarios para realizar la clasificación. Los datos de Test serán utilizados para evaluar, al final de todo, qué tan bien anda el modelo.

Para hacer esta división, utilizaremos la función [**train_test_split**](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html#sklearn.model_selection.train_test_split) de Scikit Learn.

Adicionalmente, vamos a realizar una normalización de los datos, es decir, aplicamos una operación sobre los datos de manera tal de que cada coordenada de entrada tenga aproximadamente la misma "escala" (matemáticamente, se busca que la media de cada variable de entrada sea 0 y la varianza sea unitaria). Esto se suele realizar para que no haya una variable que influya más que otra y no hayan problemas de estabilidad durante el entrenamiento.

```python
# 4. Preparamos los datos para el modelo
X_train, X_test, y_train, y_test = train_test_split(df[['x1', 'x2']], df['target'], test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
plt.figure(figsize=(5, 4))
plt.scatter(X_train['x1'], X_train['x2'], c=y_train, cmap='bwr', edgecolor='k')
plt.title('Datos de Train')
plt.colorbar(label='Valor de y')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()

plt.figure(figsize=(5, 4))
plt.scatter(X_test['x1'], X_test['x2'], c=y_test, cmap='bwr', edgecolor='k')
plt.title('Datos de Test')
plt.colorbar(label='Valor de y')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```

#### **Creación de la red**

Llegó ahora la hora que estábamos esperando: crear nuestra propia red neuronal.

Utilizaremos las siguientes funciones de Tensorflow:

- [**Sequential**](https://www.tensorflow.org/api_docs/python/tf/keras/Sequential): crea un modelo de red neuronal que organiza las capas de forma lineal, de la primera a la última, donde cada capa recibe la salida de la capa anterior y produce su propia salida para la siguiente capa.

- [**Dense**](tf.keras.layers.Dense): permite generar una capa de neuronas, cada una de las cuales recibe todas las entradas de la capa anterior. Se la llama capa densa debido a esta conexión de todas las neuronas de una capa con todas las de la capa anterior. La cantidad de salidas de esta capa densa será igual al número de neuronas.

Notemos que a la red neuronal entrarán dos entradas: la coordenada "x" y la coordenada "y" del punto. A su vez, tendrá 1 salida: la clase a la cual pertenece. Debido a que nos limitaremos a utilizar redes MLP, deberemos elegir la cantidad de capas y cantidad de neuronas de cada capa. Como ejemplo, utilizaremos capas intermedias de 8 y 4 neuronas, y una capa de salida de 1 neurona, que nos dirá la clase a la cual pertenece la predicción.

Usaremos funciones de activación "Relu" para las capas intermedias, muy utilizada por ser eficiente computacionalmente; y activación "Sigmoidea" para la capa de salida, ya que es una función de activación que mapea la salida a un valor entre 0 y 1, permitiendo realizar la clasificación a través de fijar un umbral a partir del cual la salida pertenece a una clase u a otra.

Por último, indicamos que usaremos el optimizador Adam, uno de los más utilizados, y como función de costo la "binary_crossentropy", utilizada para clasificación binaria.

Indicaremos que queremos visualizar la métrica **accuracy**, que nos indicará qué porcentaje de puntos fueron correctamente clasificados sobre el total.

```python
# 5. Definimos la red neuronal MLP
model = Sequential([
    Dense(8, input_shape=(2,), activation='relu'),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

#### **Entrenamiento de la red**

Ahora, entrenaremos al modelo con los datos de Train, utilizando el método **fit** del modelo. Deberemos indicar un conjunto de datos de Validación, para poder ir midiendo la performance del modelo a medida que va entrenando. Podemos realizar una nueva división del conjunto de Train, o bien indicarle a la función **fit** qué porcentaje del conjunto de train utilizar como validación. En este caso, elegimos un 20%.

La cantidad de **epochs** indica cuántas repeticiones del algoritmo de aprendizaje se harán sobre el conjunto total de datos de entrenamiento. El **batch size** indica la cantidad de elementos que utilizará para realizar el cálculo del gradiente y realizar la optimización. Cada **epoch** consta de varios **batches** que conforman al conjunto de train.

```python
# 6. Entrenamos el modelo
history = model.fit(X_train_scaled, y_train, batch_size=32, epochs=20, validation_split=0.2, verbose=1)
```

A continuación, graficaremos cómo evolucionó, a lo largo de los Epochs, la métrica de **Accuracy** y la función de costo.

Como puede observarse, a medida que corrieron los ephocs, el accuracy fue aumentando y la función de costo fue disminuyendo.

```python
# Accuracy
plt.figure(figsize=(6, 4))
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.title('Precisión del modelo')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Loss
plt.figure(figsize=(6, 4))
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.title('Pérdida del modelo')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

#### **Evaluación del conjunto de test**

Finalmente, corremos la función **evaluate** para correr el modelo ya entrenado sobre los datos de test y obtener el accuracy obtenido.

```python
# 7. Evaluamos
loss, acc = model.evaluate(X_test_scaled, y_test)
print(f'\n🔎 Precisión en test: {acc*100:.2f}%')
```

¡Vemos que nuestro modelo, sobre el conjunto de Test, logró clasificar correctamente al 98% de los puntos!

A continuación, las líneas de código simplemente realizan una clasificación de un gran número de puntos en el plano, de manera tal de obtener una estimación de cuál es la frontera de decisión que aprendió la red neuronal, la cual puede visualizarse en el gráfico correspondiente.

```python
# 8. Visualizamos la frontera de decisión
def plot_decision_boundary(model, X, y, scaler, title='Frontera de decisión'):
    x_min, x_max = X['x1'].min() - 1, X['x1'].max() + 1
    y_min, y_max = X['x2'].min() - 1, X['x2'].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                         np.linspace(y_min, y_max, 300))

    grid = np.c_[xx.ravel(), yy.ravel()]
    grid_scaled = scaler.transform(grid)
    probs = model.predict(grid_scaled, verbose=0).reshape(xx.shape)

    plt.figure(figsize=(7, 6))
    plt.contourf(xx, yy, probs, levels=[0, 0.5, 1], cmap="bwr", alpha=1)
    plt.scatter(X['x1'], X['x2'], c=y, cmap="bwr", edgecolor="k")
    plt.title(title)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.grid(True)
    plt.show()

plot_decision_boundary(model, df[['x1', 'x2']], df['target'], scaler)
```

En este caso, se puede ver cómo es la forntera de decisión aprendida, y cómo los puntos (en la figura se encuentran tantos los datos de Train como de Test), caen en su gran mayoría del lado predicho.

Ahora vamos a crear la red neuronal utilizando un MLP (Multilayer Perceptron). Como mencionamos antes, una imagen es una matriz de números, pero hay que recordar que la entrada a un MLP es un vector unidimensional, es decir, un conjunto de valores que, en principio, no conserva la estructura bidimensional de la imagen.

Por lo tanto, será necesario “aplanar” la imagen, transformando la matriz en un solo arreglo de valores. Esto se hace colocando una fila después de la otra, de manera que todos los píxeles queden en un vector continuo que pueda ser procesado por la red neuronal. Para realizar esto, es necesario agregar la capa **Flatten**, que realiza esta operación sobre la entrada.

Utilizaremos una red con 128 neuronas con activación Relu como primera capa, y 10 neuronas a la salida. Estas 10 neuronas representarán a los 10 dígitos/clases posibles. La idea es que cada salida represente la **probabilidad** de que cada clase sea que corresponde a la entrada dada. Para que esto sea posible, la suma de las 10 salidas debe ser **1**. Por este motivo, se utiliza la función de activación **"softmax"**, que convierte las salidas en una distribución de probabilidad sobre las clases. A la hora de la **inferencia**, se elige la clase con mayor probabilidad.

Al igual que en el caso de clasificación binaria, se usa como función de costo la **"Categorical cross-entropy"**, que generaliza a su versión binaria. También observaremos la métrica de **Accuracy**, para obtener el porcentaje de aciertos entre el total de datos.

```python
# 2. Definición del modelo
model = Sequential([
    Flatten(input_shape=(28, 28)),       # Aplana la imagen de 28x28 a 784
    Dense(128, activation='relu'),       # Capa oculta con 128 neuronas
    Dense(10, activation='softmax')      # Capa de salida con 10 clases
])

# 3. Compilación del modelo
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

Entrenamos al modelo. (Les recomendamos activar la GPU de la Notebook para correr el entrenamiento, ya que acelera notablemente el entrenamiento)

```python
# 4. Entrenamiento
history = model.fit(x_train, y_train_cat,
                    validation_split=0.2,
                    epochs=20,
                    batch_size=32)
```

```python
# Accuracy
plt.figure(figsize=(6, 4))
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.title('Precisión del modelo')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Loss
plt.figure(figsize=(6, 4))
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.title('Pérdida del modelo')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

Evaluamos para los datos de Test:

```python
# 5. Evaluación final en el set de test
test_loss, test_acc = model.evaluate(x_test, y_test_cat)
print(f"\n📈 Precisión en test: {test_acc*100:.2f}%")
```

¡Dio un 97,7% de accuracy!

## $\dificil$ Círculos concéntricos $^2$

Clasificar, utilizando una red neuronal, los puntos en el plano 2D distribuidos en forma cuatro circunferencias concéntricas. Cada sección de circunferencia pertenece a una clase distinta. Les dejamos el código que genera a los puntos y a sus respectivas clases.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

# Función auxiliar para generar un círculo con una clase
def generar_circulo(n_samples, radio, ruido, clase):
    X, _ = make_circles(n_samples=n_samples*2, factor=0.5, noise=ruido)
    # Nos quedamos solo con un círculo (el externo siempre tiene radio ~1, interno ~factor)

    X = X[_ == 0] * radio

    y = np.full(len(X), clase)
    return X, y

# Generar 4 círculos
X1, y1 = generar_circulo(1000, 0.9, 0.15, 0)   # círculo grande
X2, y2 = generar_circulo(1000, 0.6, 0.15, 1)   # círculo medio-grande
X3, y3 = generar_circulo(1000, 0.3, 0.15, 2)   # círculo medio-chico
X4, y4 = generar_circulo(1000, 0.1, 0.15, 3)   # círculo chico

# Concatenar todo
X = np.vstack([X1, X2, X3, X4])
y = np.concatenate([y1, y2, y3, y4])

# Graficar
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap="viridis", s=10)
plt.gca().set_aspect("equal")
plt.title("4 círculos concéntricos - 4 clases")
plt.show()
```

### **YoLo**

**Explicación de YOLO**

YOLO (You Only Look Once) es un algoritmo de detección de objetos basado en redes neuronales convolucionales (CNN). A diferencia de métodos tradicionales que analizan regiones específicas de una imagen, YOLO procesa la imagen completa en una sola pasada, prediciendo simultáneamente las cajas delimitadoras (bounding boxes) y las clases de los objetos detectados. Esto lo hace rápido y eficiente, ideal para aplicaciones en tiempo real como videovigilancia o conducción autónoma. YOLO divide la imagen en una cuadrícula y, para cada celda, predice coordenadas de cajas, puntajes de confianza y probabilidades de clase. Versiones recientes como YOLOv8 han mejorado su precisión y velocidad.

### **MiDaS**

**Explicación de MiDaS**

MiDaS (Monocular Depth Estimation) es un modelo de aprendizaje profundo diseñado para estimar la profundidad de una escena a partir de una sola imagen (monocular). Utiliza redes neuronales convolucionales (CNN) para predecir un mapa de profundidad, asignando a cada píxel un valor que indica su distancia relativa a la cámara. MiDaS se destaca por su capacidad de generalizar a diferentes tipos de imágenes sin necesidad de datos de profundidad específicos durante el entrenamiento, gracias a un conjunto de datos diverso. Es útil en aplicaciones como realidad aumentada, robótica y reconstrucción 3D.

##### Ejemplo de uso

```python
import cv2
import torch
import urllib.request

import matplotlib.pyplot as plt

url, filename = ("https://presenterse.com/wp-content/uploads/2020/04/ITBA.jpg", "ITBA.jpg") # Imagen
urllib.request.urlretrieve(url, filename)
```

```python
model_type = "DPT_Large"     # MiDaS v3 - Large     (highest accuracy, slowest inference speed)
#model_type = "DPT_Hybrid"   # MiDaS v3 - Hybrid    (medium accuracy, medium inference speed)
#model_type = "MiDaS_small"  # MiDaS v2.1 - Small   (lowest accuracy, highest inference speed)

midas = torch.hub.load("intel-isl/MiDaS", model_type)

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
midas.to(device)
midas.eval()

midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms")

if model_type == "DPT_Large" or model_type == "DPT_Hybrid":
    transform = midas_transforms.dpt_transform
else:
    transform = midas_transforms.small_transform
```

```python
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

input_batch = transform(img).to(device)
```

```python
with torch.no_grad():
    prediction = midas(input_batch)

    prediction = torch.nn.functional.interpolate(
        prediction.unsqueeze(1),
        size=img.shape[:2],
        mode="bicubic",
        align_corners=False,
    ).squeeze()

output = prediction.cpu().numpy()

print(output.shape, output.dtype)
```
